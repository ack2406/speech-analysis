```python
import pathlib as Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import logging

from explorations.base_exploration import BaseExploration
from explorations.gender_exploration import GenderExploration

from utils.class_utils import get_all_derived_classes
from utils.file_manager import load_csv_files_as_dataframes
from utils.logging_setup import setup_logging

setup_logging()

PATH = "../analysis_results/"
```

# Loading analyzed data

```python
data = None
if not Path.Path(PATH).exists():
    logging.error(f"Directory {PATH} does not exist")
else:
    logging.info(f"Starting to load csv files from {PATH} directory")
    data = load_csv_files_as_dataframes(PATH)
```

# Exploration

```python
exploration_classes = get_all_derived_classes(BaseExploration)

# check if classes are loaded
if len(exploration_classes) == 0:
    logging.error("No classes loaded")

for exploration_class in exploration_classes:
    exploration = exploration_class()
    exploration.initialize_data(data)
    exploration.prepare_data()
    exploration.plot_data()
```
